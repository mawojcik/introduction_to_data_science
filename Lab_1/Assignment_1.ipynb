## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
unique_grades = df['grade'].unique()

grades = [[] for _ in range(13)]

for grade in range(len(df['grade'])):
    grades[df['grade'].iloc[grade]-1].append(df['price'].iloc[grade])

plt.boxplot(grades)
plt.xlabel("grade")
plt.ylabel("price")
```

```python
#b) Create mosaic plot of any two categorical variables.

from statsmodels.graphics.mosaicplot import mosaic

#Your code:

# I eliminated not integer bathrooms numbers

df_bathroom_integer = df[df['bathrooms'] % 1 == 0]
mosaic(data=df_bathroom_integer, index=['bedrooms', 'bathrooms'], title="bedrooms")
```

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
specified_houses_df = df[(df['bedrooms'] == 3) & (df['bathrooms'] == 2) & (df['floors'] == 2)]
specified_houses_df.describe()

#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:
plt.figure(figsize=(12, 8))
plt.scatter(df['grade'], df['price'], s=4, color="purple")
plt.scatter(specified_houses_df['grade'], specified_houses_df['price'], s=40, alpha=0.5)
plt.xlabel("grade")
plt.ylabel("price")

```

```python
#c) Find such apartments with best price to footage ratio.

# I created additional column to store price to square footage ratio and then sorted by this value

#Your code:
price_to_sqft = []
for number in range(len(specified_houses_df)):
    price_to_sqft.append(specified_houses_df['price'].iloc[number] / specified_houses_df['sqft_living'].iloc[number])

specified_houses_df['price_to_sqft_ratio'] = price_to_sqft
specified_houses_df.sort_values(by=['price_to_sqft_ratio']).head(5)
#Hint: .loc, .sort_values(), .index
```
